```python
import uuid
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    'uuid': [str(uuid.uuid1()) for _ in range(5)],
    'category': ['non-food'] * 5,
    'type': ['radioactive waste'] * 5,
    'units': ['one', 'two', 'two', 'one', 'three'],
})
```

```python
df2 = pd.DataFrame({
    'uuid': [str(uuid.uuid1()) for _ in range(10)],
    'category': ['food'] * 7 + [None] * 3,
    'type': ['bamboo ice cream'] * 10,
    'units': np.random.randint(100, 500, size=10),
})
```

```python
df3 = pd.DataFrame({
    'uuid': [str(uuid.uuid1()) for _ in range(80)],
    'category': ['food'] * 80,
    'type': ['fine bamboo leaves', 'bamboo leaves (500 lbs)', 'bamboo leaves', 'Bamboo Leaves', 
             'l33t B4mb00 l34ve5', 'more bamboo leaves', 'high quality bamboo', 'leaves of the bamboo plant'] * 10,
    'units': [500] * 70 + [None] * 10,
})
```

```python
df4 = pd.DataFrame({
    'uuid': [str(uuid.uuid4()) for _ in range(20)],
    'category': ['non-food'] * 20,
    'type': ['spare gears'] * 20,
    'units': list(np.random.normal(250, 10, size=15).astype(int)) + [None] *5,
})
df4
```

```python
df = pd.concat([df1, df2, df3, df4])
df
```

```python
df = df.sample(115)
```

```python
df = df.reset_index()
del df['index']
```

```python
df.to_csv('cargo.csv')
```
