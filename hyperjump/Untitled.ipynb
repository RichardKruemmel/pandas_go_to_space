```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
np.random.seed(42)
G = np.random.normal(size=1001)
```

```python
speed = np.linspace(0, 100, 1001)
```

```python
space_fold = np.exp(speed * 0.3 + G * 100)
```

```python
time_fold = np.exp(speed * 0.4 + 100.1)
```

```python
df = pd.DataFrame({'speed': speed, 'space_fold': space_fold, 'time_fold': time_fold})
```

```python
ax = df[['time_fold', 'space_fold']].plot()
ax.set_yscale('log')
```
