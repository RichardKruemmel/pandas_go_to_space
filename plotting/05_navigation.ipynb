# Navigation

**Data Visualization**

To navigate the epsilon sector, you need an accurate and unambiguous map.
Mostly, your crew works with 2D projections of the 3D space.
The previous scatterplot was not bad, but for the big screen on the bridge
you need a well-annotated, high-resolution version of it.

Work through the starfleet standard procedure for high-quality images:

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

### Step 1: Load the data

```python
df = pd.read_csv('epsilon_sector.csv')
df.head()
```

### Step 2: Plot with default settings
Use one of the standard seaborn functions to see whether the data contains what you need.

```python
sns.scatterplot(data=df, x='x', y='y', hue='class', size='size')
```

to improve the plot, we will use `matplotlib`, a library both `pandas` plotting functions and `seaborn` are based on.

### Step 3: Format the axes
You may want to adjust the axis limits to add a bit of empty space on the sides. After all, there is plenty of empty space in space. Also, you may want to label each axis. Note that you may use *LaTeX math notation*. 

```python
sns.scatterplot(data=df, x='x', y='y', hue='class', size='size')
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.1)
plt.xticks(color="white")
plt.yticks(color="white")
plt.tick_params(color="white")
plt.xlabel("x coordinate in $\sqrt{warps}$", color="white")
plt.ylabel("y coordinate in $\sqrt{warps}$", color="#ffffff")
plt.xticks(color="white")
```

### Step 4: Add a grid

```python
sns.scatterplot(data=df, x='x', y='y', hue='class', size='size')
plt.grid()
```

### Step 5: Add points of interest
There is a black hole in Epsilon sector. It is technically not a star, so it does not appear in our data. But we should put it on the starmap with a big arrow, so that we don't accidentally get sucked into it.

```python
sns.scatterplot(data=df, x='x', y='y', hue='class', size='size')

plt.annotate('black hole $\epsilon_{23}$',
             xy=(0.6, 0.6),
             xycoords='data',
             xytext=(-90, -50),
             textcoords='offset points',
             fontsize=12,
             color="red",
             arrowprops={
                 'arrowstyle': "->",
                 'connectionstyle': "arc3,rad=.2",
                 'color': "red"
             })
```

### Step 6: Add a title
This step is crucial to understand the plot. You want everybody to be clear in which sector you are even if they overslept the last few hyperjumps.

```python
sns.scatterplot(data=df, x='x', y='y', hue='class', size='size')
plt.title('Epsilon sector x/y projection', color="white")
```

### Step 7: Set figure size
We may want a bigger image on the screen so that the officers in the back of the bridge can see everything. For historic reasons, the size of matplotlib figures (and starfleet displays) is still measured in inches.

```python
plt.figure(figsize=(11, 7))
sns.scatterplot(data=df, x='x', y='y', hue='class', size='size')
```

### Step 8: Export the image
Make the map available as an image file. Here is where you define the final resolution in pixels.
To convert from inches to pixels, the starfleet uses the ancient metric `dpi` (dots per inch):

    pixels = figure size inches * dpi

A number of image formats including png, jpg and svg are available:

```python
sns.scatterplot(data=df, x='x', y='y', hue='class', size='size')
plt.savefig('epsilon_sector.png', dpi=300)
```
